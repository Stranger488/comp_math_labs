```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
import scipy.integrate as integrate


def l_i(i, x, x_nodes):
    li = 1
    
    for j in range(len(x_nodes)):
        if j != (i - 1):
            li *= (x - x_nodes[j]) / (x_nodes[i - 1] - x_nodes[j])
    return li

def L(x, x_nodes, y_nodes):
    Lx = 0
    
    for i in range(len(x_nodes)):
        Lx += y_nodes[i] * l_i(i + 1, x, x_nodes)
    return Lx

def f(x):
    return np.exp(-x**2)

def get_cmap(n, name='winter'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)

def translate_to_interval(x, start, end):
    return 0.5 * ( (end - start) * x + start + end )

def cheb_nodes(n):
    x_cheb = np.array([])
    
    for k in range(n):
        np.append(x_cheb,
                  translate_to_interval(
                    np.cos((2 * (k + 1) - 1) * np.pi / (2 * n)),
                    -5., 5.))
    return x_cheb

def f_kus_lin(x, x_nodes, y_nodes):
    f_kus_lin = []
    a0 = 0
    a1 = 0
    
    for i in range(1, len(x_nodes)):
        if (x >= x_nodes[i - 1] and x <= x_nodes[i]):
            a1 = (y_nodes[i] - y_nodes[i - 1]) / (x_nodes[i] - x_nodes[i-1])
            a0 = y_nodes[i - 1] - a1 * x_nodes[i - 1]

            return (a0 + a1 * x_nodes[i - 1])
        
    return None
```

```python
def rem_evenly(x, x_nodes):
    n = len(x_nodes)
    
#     ksi = np.random.uniform(-5., 5.)
    ksi = 0
    
    if (n % 2 == 0):
        derivative_order = n + 1
    else:
        derivative_order = n + 2
    
    rem_evenly = derivative(f, ksi, n=n,
                            order=derivative_order) / np.math.factorial(n)
    
    for i in range(len(x_nodes)):
        rem_evenly *= (x[i] - x_nodes[i])
        
    return rem_evenly

def rem_cheb(x, x_nodes):
    n = len(x_nodes)
    
    ksi = 0
    
    if (n % 2 == 0):
        derivative_order = n + 1
    else:
        derivative_order = n + 2
    
    rem_cheb = (1 / ( 2**(n - 1) * (np.math.factorial(n)))) * derivative(f, ksi, n=n,
                            order=derivative_order) / np.math.factorial(n)
    
    return rem_cheb

def rem_kus_lin(x, x_nodes):
    n = len(x_nodes)
    
    ksi = 0
    
    if (n % 2 == 0):
        derivative_order = n + 1
    else:
        derivative_order = n + 2
    
    h = ( (-5) - (5) ) / (n - 1)
    
    rem_kus_lin = (np.e * h**2) / 8 
        
    return rem_kus_lin
```

```python
x = np.linspace(-5., 5., 200, endpoint=True)
fx = f(x)
N = np.arange(4, 21)

cmap = get_cmap(len(N))

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6), dpi=100)
axes.plot(x, fx, color="red", linewidth=2.0)

for n in N:
    x_nodes_evenly_distributed = np.linspace(-5., 5., n)
    axes.plot(x, L(x, x_nodes_evenly_distributed, f(x_nodes_evenly_distributed)),
              color=cmap(n - 4), linestyle='--', label=str(n), linewidth=1.0) 

axes.legend(loc='upper left')
axes.grid()

plt.show()
```

```python
norm_evenly = []
norm_cheb = []
norm_kus_lin = []

rem_evenly_array = []
rem_cheb_array = []
rem_kus_lin_array = []

for n in N:
    x_nodes_evenly_distributed = np.linspace(-5., 5., n)
    x_nodes_cheb = cheb_nodes(n)
    
    y_L_evenly = []
    y_cheb = []
    y_kus_lin = []
    
    diff_evenly = []
    diff_cheb = []
    diff_kus_lin = []
       
    for i in range(len(x)):
        y_L_evenly.append(L(x[i], x_nodes_evenly_distributed, f(x_nodes_evenly_distributed)))
        y_cheb.append(L(x[i], x_nodes_cheb, f(x_nodes_cheb)))
        y_kus_lin.append(f_kus_lin(x[i], x_nodes_evenly_distributed, f(x_nodes_evenly_distributed)))
        
        diff_evenly.append(np.abs(f(x[i]) - y_L_evenly[i]))
        diff_cheb.append(np.abs(f(x[i]) - y_cheb[i]))
        diff_kus_lin.append(np.abs(f(x[i]) - y_kus_lin[i]))
        
    norm_evenly.append(max(diff_evenly))
    norm_cheb.append(max(diff_cheb))
    norm_kus_lin.append(max(diff_kus_lin))
    
    rem_evenly_array.append(rem_evenly(x, x_nodes_evenly_distributed))
    rem_cheb_array.append(rem_cheb(x, x_nodes_cheb))
    rem_kus_lin_array.append(rem_kus_lin(x, x_nodes_evenly_distributed))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6), dpi=100)
axes.semilogy(N, norm_evenly, '--o', color="red", linewidth=1.0, label="равномерное распределение")
axes.semilogy(N, rem_evenly_array, '--o', color="blue",
              linewidth=1.0, label="остаточный член, равномерное распределение")
axes.grid()
axes.legend(loc='lower left')
plt.show()


fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6), dpi=100)
axes.semilogy(N, norm_cheb, '--o', color="green", linewidth=1.0, label="оптимальное распределение")
axes.semilogy(N, rem_cheb_array, '--o', color="purple",
              linewidth=1.0, label="остаточный член, оптимальное распределение")
axes.grid()
axes.legend(loc='lower left')
plt.show()


fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6), dpi=100)
axes.semilogy(N, norm_kus_lin, '--o', color="orange", linewidth=1.0, label="кусочно-линейная интерполяция")
axes.semilogy(N, rem_kus_lin_array, '--o', color="grey",
              linewidth=1.0, label="остаточный член, кусочно-линейная интерполяция")
axes.grid()
axes.legend(loc='lower left')
plt.show()
```
